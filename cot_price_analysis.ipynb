```python
df_new = df_combined[['Market and Exchange Names', 
'As of Date in Form YYYY-MM-DD',
'Open Interest (All)',
'Noncommercial Positions-Long (All)',
'Noncommercial Positions-Short (All)',
'Noncommercial Positions-Spreading (All)',
'Commercial Positions-Long (All)',
'Commercial Positions-Short (All)',
# 'Total Reportable Positions-Long (All)',
# 'Total Reportable Positions-Short (All)',
'Nonreportable Positions-Long (All)',
'Nonreportable Positions-Short (All)'
 ]]
   
```

```python
df_frame = df_new
```

```python
df_frame.set_index('As of Date in Form YYYY-MM-DD', inplace=True)
```

```python
import plotly.express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots
```

```python
df_frame['Net_Position_Non_Commercial'] = df_frame['Noncommercial Positions-Long (All)'] - df_frame['Noncommercial Positions-Short (All)']
df_frame['Net_Position_Commercial'] = df_frame['Commercial Positions-Long (All)'] - df_frame['Commercial Positions-Short (All)']
df_frame['Net_Position_nonreportable'] = df_frame['Nonreportable Positions-Long (All)'] - df_frame['Nonreportable Positions-Short (All)']
```

```python
eurofx_cot = df_frame[(df_frame['Market and Exchange Names'].str.startswith('EURO FX -'))]
cadusd_cot = df_frame[(df_frame['Market and Exchange Names'].str.startswith('CANADIAN DOLLAR -'))]
audusd_cot = df_frame[(df_frame['Market and Exchange Names'].str.startswith('AUSTRALIAN DOLLAR -'))]
chfusd_cot = df_frame[(df_frame['Market and Exchange Names'].str.startswith('SWISS FRANC -'))]
jpyusd_cot = df_frame[(df_frame['Market and Exchange Names'].str.startswith('JAPANESE YEN -'))]
mxnusd_cot = df_frame[(df_frame['Market and Exchange Names'].str.startswith('MEXICAN PESO -'))]
gold_cot = df_frame[(df_frame['Market and Exchange Names'].str.startswith('GOLD -'))]
silver_cot = df_frame[(df_frame['Market and Exchange Names'].str.startswith('SILVER -'))]
usd_index_cot = df_frame[(df_frame['Market and Exchange Names'].str.startswith('U.S. DOLLAR INDEX -')) |
                         (df_frame['Market and Exchange Names'].str.startswith('USD INDEX -')) ]
gbpusd_cot = df_frame[(df_frame['Market and Exchange Names'].str.startswith('BRITISH POUND STERLING -')) |
                         (df_frame['Market and Exchange Names'].str.startswith('POUND STERLING -')) ]
nzdusd_cot = df_frame[(df_frame['Market and Exchange Names'].str.startswith('NEW ZEALAND DOLLAR')) |
                         (df_frame['Market and Exchange Names'].str.startswith('NZ DOLLAR -')) ]
crudeoil_cot = df_frame[(df_frame['Market and Exchange Names'].str.startswith('CRUDE OIL'))]
sp500_cot = df_frame[(df_frame['Market and Exchange Names'].str.startswith('STOCK INDEX FUTURE, S&P 500')) |
                         (df_frame['Market and Exchange Names'].str.startswith('S&P 500 STOCK INDEX -')) ]
```

```python
fig = px.area(gold_cot, x=gold_cot.index, y=['Net_Position_Non_Commercial','Net_Position_Commercial','Net_Position_nonreportable'],
              title='Net Position of Non-Commercial Traders in Gold Futures and Options',
              labels={'As of Date in Form YYYY-MM-DD': 'Date', 'Net_Position_Non_Commercial': 'Net Position'})
fig.update_layout(xaxis_title='Date', yaxis_title='Net Position')
fig.show()
```

```python

# Create the candlestick chart for the first subplot
fig1 = go.Figure(data=[go.Candlestick(x=df_fx['time'],
                                      open=df_fx['open'],
                                      high=df_fx['high'],
                                      low=df_fx['low'],
                                      close=df_fx['close'],
                                      name='FX Prices')])
fig1.update_layout(xaxis_title='Date', yaxis_title='Net Position')

fig1.update_layout(title='FX Prices Candlestick Chart')

# Create the subplot figure with two rows
fig = make_subplots(rows=2, cols=1, shared_xaxes=True,
                    subplot_titles=('FX Prices Candlestick Chart', 'Net Positions of Traders'),
                    vertical_spacing=0.02, row_heights=[0.8, 0.2])

# Add the candlestick chart to the first row
for trace in fig1['data']:
    fig.add_trace(trace, row=1, col=1)

# Create the second plot with bar charts for net positions
fig2 = go.Figure()
fig2.add_trace(go.Bar(x=gold_cot.index, y=gold_cot['Net_Position_Non_Commercial'], name='Non-Commercial'))
fig2.add_trace(go.Bar(x=gold_cot.index, y=gold_cot['Net_Position_Commercial'], name='Commercial'))
fig2.add_trace(go.Bar(x=gold_cot.index, y=gold_cot['Net_Position_nonreportable'], name='Non-Reportable'))

fig2.update_layout(barmode='group')

# Add the bar charts to the second row
for trace in fig2['data']:
    fig.add_trace(trace, row=2, col=1)

# Update layout to share cursor and include a range slider
fig.update_layout(
    height=800,
    title_text='Combined Plot with Subplots',
    showlegend=True,
    hovermode='x unified',  # This setting shares the cursor along the x-axis
    plot_bgcolor='rgba(30,30,30,1)',  # Change the plot background color
    paper_bgcolor='rgba(30,30,30,1)',  # Change the paper background color
    font=dict(color='white'),  # Change the font color to white for better contrast
    xaxis=dict(
        showspikes=True,
        spikemode='across',
        spikesnap='cursor',
        showline=True,
        rangeslider=dict(visible=False),
        rangeselector=dict(
            buttons=list([
                dict(count=1, label="1m", step="month", stepmode="backward"),
                dict(count=6, label="6m", step="month", stepmode="backward"),
                dict(count=1, label="YTD", step="year", stepmode="todate"),
                dict(count=1, label="1y", step="year", stepmode="backward"),
                dict(step="all")
            ])
        )
    ),
    xaxis2=dict(
        showspikes=True,
        spikemode='across',
        spikesnap='cursor',
        showline=True,
        rangeslider=dict(visible=False)  # Add the range slider
    ),
    yaxis=dict(showgrid=True, gridcolor='rgba(255,255,255,0.2)'),  # Lighter grid lines
    yaxis2=dict(showgrid=True, gridcolor='rgba(255,255,255,0.2)')  # Lighter grid lines
)

# Show the combined plot with subplots
fig.show()
```
